```python
import pandas as pd
import os
path = './sales_2019'
files = [file for file in os.listdir(path) if not file.startswith('.')]
all_months_data = pd.DataFrame()
for file in files:
    curr_files = pd.read_csv(path+"/"+file)
    all_months_data = pd.concat([all_months_data,curr_files])
all_months_data
```

```python
all_months_data.to_csv("yearly_sales_2019.csv",index=False)
```

```python
df = pd.read_csv("yearly_sales_2019.csv")
df
```

```python
df.head()
```

```python
re_nan = df[df.isna().any(axis=1)]
display(re_nan)
```

```python
df = df.dropna(how='all')
```

```python
df.head()
```

```python
df = df[df['Order Date'].str[0:2]!='Or']
```

```python
df.head()
```

```python
df['Month'] = pd.to_datetime(df['Order Date']).dt.month
df.head()
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
```

```python
df.head()
```

```python
def get_city(address):
    return address.split(',')[1].strip(" ")
def get_code(address):
    return address.split(',')[2].split(' ')[1]

df['City'] = df['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_code(x)})")
df.head()
```

### Which month got highest sales?

```python
df['Sales'] = df['Quantity Ordered'].astype('int') * df['Price Each'].astype('float')
```

```python
df.head()
```

```python
df.groupby(['Month']).sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months,df.groupby(['Month']).sum()['Sales'])

plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

### Which city got highest product sales

```python
df.groupby(['City']).sum()['Sales']
```

```python
cities = [city for city, df in df.groupby(['City'])]
sales = df.groupby(['City']).sum()['Sales']

plt.bar(cities,sales)
plt.xlabel('Cities')
plt.xticks(cities,rotation='vertical',fontsize=10)
plt.ylabel('Sales in USD($)')
plt.show()
```

### Conclusion : SanFranciso got highest product Sales

### Question 3: What time should we display advertisements to maximize likelihood of customer's buying product?

```python
df.head(10)
```

```python
df['Hour'] = pd.to_datetime(df['Order Date']).dt.hour
df['Minute'] = pd.to_datetime(df['Order Date']).dt.minute
df['Count'] = 1
df.head(10)
```

```python
keys = [pair for pair,df in df.groupby(['Hour'])]
values = df.groupby(['Hour']).sum()['Count']
plt.plot(keys,values)
plt.xlabel('Hours')
plt.ylabel('No.of Products')
plt.xticks(keys)
plt.grid()
plt.show()
```

### Question 4: What products are most often sold together?

```python
new_df = df.groupby(['Order ID']).filter(lambda x : len(x) > 1)
new_df
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df2 = df[['Order ID', 'Grouped']].drop_duplicates()
df2
```

```python

from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```

```python
product_group = df.groupby(['Product'])
quantity_ordered = product_group.sum()['Quantity Ordered']
keys = [pair for pair,df in product_group]
plt.bar(keys,quantity_ordered)
plt.xticks(keys,rotation='vertical',size=10)
plt.show()
```

```python
# Referenced: https://stackoverflow.com/questions/14762181/adding-a-y-axis-label-to-secondary-y-axis-in-matplotlib

prices = df.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, quantity_ordered, color='g')
ax2.plot(keys, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(keys, rotation='vertical', size=8)

fig.show()
```
