```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('loan2.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df['Education'].unique()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
```

```python
df['Married'].fillna(df['Married'].mode()[0],inplace=True)
```

```python
df['Dependents']=df['Dependents'].map({
    '0':'0',
    '1':'1',
    '2':'2',
    '3+':'3'
})
```

```python
df['Dependents'].unique()
```

```python
df.dropna(subset=['Dependents'],inplace=True)
```

```python
df['Self_Employed'].value_counts()
```

```python
df.dropna(subset=['Self_Employed'],inplace=True)
```

```python
df['LoanAmount'].fillna(df['LoanAmount'].mean(),inplace=True)
```

```python
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean(),inplace=True)
```

```python
df['Credit_History'].value_counts(normalize=True)
```

```python
df['Credit_History'].fillna(df['Credit_History'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.describe()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Loan_Status', data=df)
plt.title('Loan Status Distribution')
plt.xlabel('Loan Status')
plt.ylabel('Count')
plt.show()
```

```python
num_cols = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History']
plt.figure(figsize=(8,6))
for i,col in enumerate(num_cols,1):
    plt.subplot(3,2,i)
    sns.histplot(df[col],bins=20,kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
df.columns
```

```python
cat_cols = ['Gender','Married', 'Dependents', 'Education','Self_Employed','Property_Area','Credit_History']
plt.figure(figsize=(16,16))
for i,col in enumerate(cat_cols,1):
    plt.subplot(4,2,i)
    sns.countplot(x=col,data=df)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
num_cols=num_cols = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Loan_Status']
plt.figure(figsize=(12, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x='Loan_Status', y=col, data=df)
    plt.title(f'Box Plot of {col} by Loan Status')
    plt.xlabel('Loan Status')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```
